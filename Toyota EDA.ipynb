 # Dealing with Missing Values


```python
df1.isna().sum()
```

```python
df1.info()
```

```python
df1.describe()
```

```python
df1['Age'].fillna(df1['Age'].median(),inplace = True)
df1['KM'].fillna(df1['KM'].median(),inplace = True)
df1['HP'].fillna(df1['HP'].median(),inplace = True)
df1['MetColor'].fillna(df1['MetColor'].median(),inplace = True)
```

```python
df1.isna().sum()
```

```python
df1
```

# for loop

```python
for i in range(0,len(df1['Price']),1):
    if df1['Price'][i] <= 8450:
        df1.loc[i, 'Price_Class'] = 'Low'
    elif df1['Price'][i] > 11950:
        df1.loc[i, 'Price_Class'] = 'High'
    else:
        df1.loc[i, 'Price_Class'] = 'Medium'
# df[col_name][index_no]<>= :
# df.loc[i,"new_col"] = ""
df1
```

# While Loop 

```python
i =0
while i < len(df1['Price']):
    if df1['Price'][i] <= 8450:
        df1.loc[i, 'Price_Class'] = 'Low'
    elif df1['Price'][i] > 11950:
        df1.loc[i, 'Price_Class'] = 'High'
    else:
        df1.loc[i, 'Price_Class'] = 'Medium'
    i = i+1
df1
```

# performing exploratory data analysis(EDA)

```python
df2 = df.copy()
df2
```

```python
# frequency table
pd.crosstab(index = df2['FuelType'],columns = 'count',dropna = True)
```

```python
# Two - Way table
pd.crosstab(index=df2['Automatic'],columns=df2['FuelType'],dropna=True)
```

```python
# Two - Way table - joint probability 
pd.crosstab(index=df2['Automatic'],columns=df2['FuelType'],normalize = True ,dropna=True)
```

```python
# Two - Way table - marginal probability 
pd.crosstab(index=df2['Automatic'],columns=df2['FuelType'],margins = True,normalize = True ,dropna=True)
```

```python
# Two - Way table - conditional probability 
pd.crosstab(index=df2['Automatic'],columns=df2['FuelType'],margins = True,normalize = 'index' ,dropna=True)
```
