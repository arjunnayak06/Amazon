 # Changing Data Types of Columns from object to float

```python
# Changing the data type of discounted price and actual price

df['discounted_price'] = df['discounted_price'].str.replace("₹",'')
df['discounted_price'] = df['discounted_price'].str.replace(",",'')
df['discounted_price'] = df['discounted_price'].astype('float64')

df['actual_price'] = df['actual_price'].str.replace("₹",'')
df['actual_price'] = df['actual_price'].str.replace(",",'')
df['actual_price'] = df['actual_price'].astype('float64')
```

```python
# # Changing Datatype and values in Discount Percentage
df['discount_percentage'] = df['discount_percentage'].str.replace("%",'')
df['discount_percentage'] = df['discount_percentage'].astype('float64')
df['discount_percentage'] = df['discount_percentage']/100
```

```python
# Finding unusual string in rating column
df['rating'].value_counts()
```

```python
# Check the strange row
df.query('rating == "|"')
```

```python
#  changing the rating_count data type to float64
df['rating_count'] = df['rating_count'].str.replace(",",'')
df['rating_count'] = df['rating_count'].astype('float64')
```

```python
# changing rating column data type to float64
df['rating'] = df['rating'].str.replace('|','3.9')
df['rating'] = df['rating'].astype('float64')
```

# Statistical Tests

```python
import scipy.stats as stats
```

```python
# Conduct t-test to compare rating between two categories
t_statistic,p_value = stats.ttest_ind(df[df['category'] == 'electronics']['rating'],df[df['category'] == 'clothing']['rating'])
print(t_statistic,p_value)
```

```python
df.info()
```

```python
# Chi-square test
# Create a contigency table

contigency_table = pd.crosstab(df['actual_price'],df['rating'])
contigency_table 

# A contingency table is a data table that compares two variables. In the contingency table,
# one data set is populated in rows, while the other data set is populated in columns. 
# Values of the cells where the rows and columns intersect can suggest whether or not the two sets are correlated.
```

```python
# perform chi-square test
chi2, p, dof, expected = stats.chi2_contingency(contigency_table)
print('Chi-square statistic:', chi2)
print('p-value:', p)
print('Degrees of freedom:', dof)
print(f"Expected:\n {expected}")
```
